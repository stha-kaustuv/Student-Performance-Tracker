```python
#find all string columns
string_columns_name=[]
for col in df.columns:
    if df[col].dtype == 'str':
        string_columns_name.append(col)
        print(col)
```

```python
#find unique item and their value counts of str-datatype columns
for col in string_columns_name:
    print(f'{col}')
    print('....................................')
    print(df[col].unique())
    print(df[col].value_counts())
    print()
```
